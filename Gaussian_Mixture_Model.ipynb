```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# generating synthetic datset
np.random.seed(20)
X, Y_true = make_blobs(n_samples = 1000, centers = 5, cluster_std = [0.5, 1.55, 2, 3.6, 3], random_state = 20)
# x means the data of the data point
# Y_true means the actual cluster to which the data point  should belong
```

```python
# Kmeans to do clustering
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 20)
kmeans_labels = kmeans.fit_predict(X)
kmeans_labels
```

```python
# GMM clustering
```

```python
gmm = GaussianMixture(n_components = 5, random_state = 20)
gmm_labels = gmm.fit_predict(X)
gmm_labels
```

```python
# comparing the output by visual comparision
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 5), sharex = True, sharey = True)
axes[0].scatter(X[:, 0], X[:, 1], c = Y_true, cmap = 'Dark2' , s = 40)
axes[0].set_title('True cluster')

axes[1].scatter(X[:, 0], X[:, 1], c = kmeans_labels, cmap = 'viridis' , s = 40)
axes[1].set_title('KMeans cluster')

axes[2].scatter(X[:, 0], X[:, 1], c = gmm_labels, cmap = 'coolwarm' , s = 40)
axes[2].set_title('GMM cluster')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_kmeans = silhouette_score(X, kmeans_labels)
silhouette_gmm = silhouette_score(X, gmm_labels)

print(f"Kmeans Score : {silhouette_kmeans}")
print(f"GMM Score : {silhouette_gmm}")
```
